## 1. Load Raw Data

````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load raw files
df_m_raw = pd.read_excel("data/raw/mamaearth_financials_raw.xlsx")
df_b_raw = pd.read_excel("data/raw/boat_financials_raw.xlsx")

print("Mamaearth raw shape:", df_m_raw.shape)
print("boAt raw shape:", df_b_raw.shape)
print("\nMamaearth columns:", list(df_m_raw.columns))
````

## 2. Clean & Standardize

````python
def clean_dataframe(df, brand_name):
    """Standardize column names, add brand tag, verify fiscal year format."""
    df = df.copy()

    # Drop non-numeric metadata columns (like ipo_status for boAt)
    drop_cols = [c for c in df.columns if df[c].dtype == object and c != 'fiscal_year']
    if drop_cols:
        print(f"  Dropping non-numeric metadata: {drop_cols}")
        df = df.drop(columns=drop_cols)

    # Standardize fiscal year
    df['fiscal_year'] = df['fiscal_year'].str.upper().str.strip()

    # Add brand identifier
    df.insert(0, 'brand', brand_name)

    # Verify no nulls in critical columns
    critical = ['fiscal_year', 'revenue_cr', 'gross_margin_pct', 'marketing_spend_cr']
    null_check = df[critical].isnull().sum()
    if null_check.any():
        print(f"  ⚠️  Nulls in critical columns:\n{null_check[null_check > 0]}")
    else:
        print(f"  ✅ No nulls in critical columns")

    # Validate: gross margin must be between 0-100
    assert df['gross_margin_pct'].between(0, 100).all(), "Gross margin out of range"
    print(f"  ✅ Gross margin range valid: {df['gross_margin_pct'].min():.1f}% – {df['gross_margin_pct'].max():.1f}%")

    return df

print("Cleaning Mamaearth...")
df_m = clean_dataframe(df_m_raw, "Mamaearth")

print("\nCleaning boAt...")
df_b = clean_dataframe(df_b_raw, "boAt")
````

## 3. Derive Unit Economics Metrics

### Derivation Formulas

| Metric | Formula | Confidence |
|--------|---------|------------|
| **CAC** | Marketing Spend ÷ Estimated New Customers | Medium — new customers derived |
| **LTV** | AOV × Orders/Year × Gross Margin% × Customer Lifespan | Medium — lifespan estimated |
| **LTV:CAC** | LTV ÷ CAC | Derived from above |
| **Payback Period** | CAC ÷ (AOV × Gross Margin% × Orders/Year) | Medium |
| **Marketing Efficiency Ratio** | Revenue ÷ Marketing Spend | High — directly reported |
| **CAC Intensity** | Marketing Spend ÷ Revenue | High — directly reported |

> **CAC Derivation Note:** Neither brand directly discloses CAC. We derive it as Marketing Spend ÷ Estimated New Customer Additions. New customer additions are derived from YoY delta in active consumer base + estimated 18-22% annual churn (Mamaearth) and 25-30% (boAt). This methodology is documented in `methodology.md` and cross-verified against ICICI Securities analyst estimates where available.

````python
def derive_unit_economics(df):
    """
    Derive all unit economics metrics.
    Every formula documented with assumption and confidence level.
    """
    df = df.copy()

    # ── CAC ──────────────────────────────────────────────────────────────────
    # Formula: Marketing Spend (₹Cr) / New Customers (Mn) × 10 → ₹ per customer
    # Marketing Spend in ₹Cr, New Customers in millions
    # Result: CAC in ₹ (rupees per new customer)
    # Assumption: All marketing spend contributes to new customer acquisition (conservative)
    # Confidence: MEDIUM — new_customers_mn is derived, not directly reported
    df['cac_rs'] = (df['marketing_spend_cr'] * 1e7) / (df['new_customers_mn'] * 1e6)
    # = ₹Cr × 10,000,000 / (Mn × 1,000,000) = ₹ per customer

    # ── LTV ──────────────────────────────────────────────────────────────────
    # Formula: AOV × Orders/Year × Gross Margin% × Customer Lifespan (years)
    # Gross Margin expressed as decimal
    # Confidence: MEDIUM — customer lifespan estimated from industry benchmarks
    df['ltv_rs'] = (df['avg_order_value_rs']
                    * df['avg_orders_per_customer_yr']
                    * (df['gross_margin_pct'] / 100)
                    * df['avg_customer_lifespan_yrs'])

    # ── LTV:CAC Ratio ─────────────────────────────────────────────────────────
    # Threshold: 3x = critical minimum | 5x = healthy | >6x = exceptional
    df['ltv_cac_ratio'] = df['ltv_rs'] / df['cac_rs']

    # ── Payback Period (months) ───────────────────────────────────────────────
    # Formula: CAC / (AOV × Gross Margin% × Orders/Year) × 12
    # = months to recover CAC from gross profit contribution
    df['payback_period_months'] = (
        df['cac_rs'] / (
            df['avg_order_value_rs']
            * (df['gross_margin_pct'] / 100)
            * df['avg_orders_per_customer_yr']
        )
    ) * 12

    # ── Marketing Efficiency Ratio ────────────────────────────────────────────
    # Revenue generated per ₹1 of marketing spend = proxy for blended ROAS
    # Higher = better; benchmark target 3-5x
    df['marketing_efficiency_ratio'] = df['revenue_cr'] / df['marketing_spend_cr']

    # ── CAC Intensity (%) ─────────────────────────────────────────────────────
    # Marketing Spend as % of Revenue — how much revenue goes back into acquisition
    df['cac_intensity_pct'] = (df['marketing_spend_cr'] / df['revenue_cr']) * 100

    # ── CAC YoY Change (%) ────────────────────────────────────────────────────
    df['cac_yoy_change_pct'] = df.groupby('brand')['cac_rs'].pct_change() * 100

    # ── LTV YoY Change (%) ────────────────────────────────────────────────────
    df['ltv_yoy_change_pct'] = df.groupby('brand')['ltv_rs'].pct_change() * 100

    # ── Revenue CAGR basis ────────────────────────────────────────────────────
    df['revenue_yoy_growth_pct'] = df.groupby('brand')['revenue_cr'].pct_change() * 100

    return df

df_m = derive_unit_economics(df_m)
df_b = derive_unit_economics(df_b)

print("✅ Unit economics derived for both brands")
print("\n── Mamaearth LTV:CAC ──")
print(df_m[['fiscal_year','cac_rs','ltv_rs','ltv_cac_ratio','payback_period_months','marketing_efficiency_ratio']].round(2).to_string(index=False))
print("\n── boAt LTV:CAC ──")
print(df_b[['fiscal_year','cac_rs','ltv_rs','ltv_cac_ratio','payback_period_months','marketing_efficiency_ratio']].round(2).to_string(index=False))
````

## Setup

````python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os, warnings
warnings.filterwarnings('ignore')

os.makedirs("outputs/charts", exist_ok=True)

# Load master dataset
master = pd.read_csv("data/clean/master_metrics.csv")
df_m = master[master['brand'] == 'Mamaearth'].reset_index(drop=True)
df_b = master[master['brand'] == 'boAt'].reset_index(drop=True)

# Brand colors
COLORS = {
    'Mamaearth': '#27ae60',   # Green
    'boAt':      '#e74c3c',   # Red
    'benchmark': '#f39c12',   # Orange for benchmark lines
    'danger':    '#c0392b',   # Dark red for danger zones
    'positive':  '#2ecc71',
    'bg':        '#1a1a2e',
    'text':      '#ecf0f1',
}

TEMPLATE = "plotly_dark"

print("✅ Setup complete")
print(f"Dataset: {master.shape[0]} rows × {master.shape[1]} cols")
print(f"Brands: {master['brand'].unique()}")
print(f"Years:  {master['fiscal_year'].unique()}")
````

## CHART 0: Unit Economics Engine

````python
# ─── CHART 0: Unit Economics Engine ─────────────────────────────

# ─── Step 1: Detect / Create Required Columns ───

def find_col(possible_names):
    for col in master.columns:
        for name in possible_names:
            if name.lower() in col.lower():
                return col
    return None

CAC_COL = find_col(['cac', 'acquisition'])
LTV_COL = find_col(['ltv', 'lifetime'])
RATIO_COL = find_col(['ltv_cac', 'ratio'])

# ─── Step 2: Create metrics if missing ───
if CAC_COL is None:
    if 'marketing_spend' in master.columns and 'new_customers' in master.columns:
        master['cac'] = master['marketing_spend'] / master['new_customers']
        CAC_COL = 'cac'

if LTV_COL is None:
    if all(col in master.columns for col in ['avg_order_value', 'purchase_frequency', 'gross_margin']):
        master['ltv'] = (
            master['avg_order_value'] *
            master['purchase_frequency'] *
            master['gross_margin']
        )
        LTV_COL = 'ltv'

if RATIO_COL is None and CAC_COL and LTV_COL:
    master['ltv_cac_ratio'] = master[LTV_COL] / master[CAC_COL]
    RATIO_COL = 'ltv_cac_ratio'

print("Using columns →", CAC_COL, LTV_COL, RATIO_COL)


# ─── Step 3: Create Chart ───

fig0 = make_subplots(
    rows=1, cols=2,
    subplot_titles=("CAC vs LTV (₹ per Customer)", "LTV:CAC Ratio vs 3x Benchmark"),
    horizontal_spacing=0.12
)

# ─── CAC vs LTV ───
for brand, color in [('Mamaearth', COLORS['Mamaearth']), ('boAt', COLORS['boAt'])]:
    df_ = master[master['brand'] == brand]

    # CAC
    fig0.add_trace(go.Scatter(
        x=df_['fiscal_year'], y=df_[CAC_COL],
        mode='lines+markers',
        name=f"{brand} CAC",
        line=dict(color=color, width=2, dash='dot'),
        marker=dict(size=8),
        legendgroup=brand
    ), row=1, col=1)

    # LTV
    fig0.add_trace(go.Scatter(
        x=df_['fiscal_year'], y=df_[LTV_COL],
        mode='lines+markers',
        name=f"{brand} LTV",
        line=dict(color=color, width=3),
        marker=dict(size=10),
        legendgroup=brand
    ), row=1, col=1)


# ─── LTV:CAC Ratio ───
for brand, color in [('Mamaearth', COLORS['Mamaearth']), ('boAt', COLORS['boAt'])]:
    df_ = master[master['brand'] == brand]

    fig0.add_trace(go.Bar(
        x=df_['fiscal_year'],
        y=df_[RATIO_COL],
        name=brand,
        marker_color=color,
        opacity=0.85,
        text=[f"{v:.1f}x" for v in df_[RATIO_COL]],
        textposition='outside',
        showlegend=False
    ), row=1, col=2)


# ─── 3x Benchmark Line ───
fig0.add_hline(
    y=3,
    line_dash="dash",
    line_color="#e74c3c",
    annotation_text="3x Sustainability Threshold",
    annotation_position="top right",
    row=1, col=2
)


# ─── Layout ───
fig0.update_layout(
    title=dict(
        text="<b>Unit Economics Engine — CAC vs LTV Sustainability (FY22–FY25)</b><br><sup>Key Investor Metric: LTV:CAC ≥ 3x</sup>",
        x=0.5,
        font=dict(size=16)
    ),
    template=TEMPLATE,
    height=420,
    legend=dict(orientation="h", y=1.12, x=0.5, xanchor='center')
)


# ─── Insight ───
fig0.add_annotation(
    text="📌 <b>Core Insight:</b> LTV:CAC compression indicates rising acquisition inefficiency — watch CAC inflation vs retention gains.",
    xref="paper", yref="paper",
    x=0.5, y=-0.15,
    showarrow=False,
    font=dict(size=11, color="#e74c3c"),
    bgcolor="rgba(231,76,60,0.1)",
    bordercolor="#e74c3c",
    borderwidth=1,
    xanchor='center'
)


# ─── Save + Show ───
fig0.write_html("outputs/charts/00_unit_economics_engine.html")
fig0.show()

print("✅ Chart 0 (Hero) added successfully")
````

This chart presents a consolidated view of **unit economics sustainability** for Mamaearth and boAt across FY22–FY25 by comparing customer acquisition costs (CAC), lifetime value (LTV), and the resulting LTV:CAC ratio.

---

### 1. CAC vs LTV (Left Panel)

The left side of the chart compares how much each company spends to acquire a customer (CAC) versus the value generated from that customer over time (LTV).

* **Mamaearth** shows a steady increase in LTV from FY22 to FY25, indicating improving customer value through higher repeat purchases or better monetization. However, CAC also rises over time, especially toward FY25, narrowing the gap between LTV and CAC.
* **boAt** operates at a significantly higher CAC level compared to Mamaearth, reflecting aggressive marketing or customer acquisition strategies. Its LTV also increases but not proportionally enough to create a consistently widening margin over CAC.

The key takeaway from this panel is the **gap between LTV and CAC**:

* A wider gap indicates stronger unit economics.
* A narrowing gap suggests declining efficiency in customer acquisition.

For Mamaearth, this gap begins to compress in later years, signaling emerging pressure. For boAt, the gap remains positive but volatile.

---

### 2. LTV:CAC Ratio vs Benchmark (Right Panel)

The right side translates the relationship into a more interpretable efficiency metric: the **LTV:CAC ratio**, with a benchmark line at **3x**, commonly considered the minimum threshold for sustainable growth in D2C businesses.

* **Mamaearth** starts at a very low ratio (~1x in FY22), improves to around 1.4x in FY24, but declines again in FY25. This indicates that despite growth, the business is **not consistently meeting the 3x sustainability threshold**, making its growth potentially inefficient.
* **boAt** maintains ratios above 3x across all years, peaking near 4.7x. This suggests stronger unit economics, although the fluctuations indicate some instability in maintaining efficiency.

---

### 3. Overall Interpretation

The chart highlights a critical contrast:

* **Mamaearth** is growing but at the cost of increasing acquisition expenses relative to customer value. The declining or sub-threshold LTV:CAC ratio indicates **risk of margin compression and unsustainable growth if trends continue**.

* **boAt**, while spending heavily on acquisition, still maintains LTV levels sufficient to keep its ratio above the sustainability benchmark. This suggests **better monetization or retention efficiency**, though variability indicates sensitivity to marketing spend changes.

---

### 4. Strategic Insight

The core insight from this visualization is that **growth alone is not sufficient**. The sustainability of that growth depends on maintaining a healthy balance between CAC and LTV.

* If CAC continues to rise without proportional LTV expansion, profitability will deteriorate.
* Businesses must either improve retention, increase average order value, or optimize marketing efficiency to maintain or expand the LTV:CAC ratio.

This chart effectively reframes the analysis from revenue growth to **quality of growth**, which is the primary concern for investors and operators evaluating long-term scalability.

## Core Scenario Engine

| Scenario | CAC Change | LTV Change | Revenue Growth | Rationale |
|----------|------------|------------|---------------|----------|
| **Bear** | +30% YoY | -5% YoY | +5% YoY | Matches 2022-2024 actual CAC inflation; ROAS below 3x |
| **Base** | +15% YoY | +2% YoY | +12% YoY | Moderated inflation; partial efficiency gains |
| **Bull** | -10% YoY | +8% YoY | +18% YoY | Retention investment + organic channel expansion |

````python
def project_scenario(brand_name, base_vals, cac_change_pct, ltv_change_pct,
                     revenue_growth_pct, years=3):
    """
    Project LTV:CAC ratio forward under given scenario assumptions.

    Parameters:
    -----------
    brand_name       : str
    base_vals        : dict — FY25 actuals
    cac_change_pct   : float — annual % change in CAC (e.g., +30 = 30% inflation)
    ltv_change_pct   : float — annual % change in LTV
    revenue_growth_pct: float — annual % revenue growth (for context)
    years            : int — projection horizon

    Returns:
    --------
    DataFrame with year-by-year projections
    """
    rows = []
    cac = base_vals['cac_rs']
    ltv = base_vals['ltv_rs']
    rev = base_vals['revenue_cr']

    for yr in range(1, years + 1):
        fy = f"FY{25 + yr}"
        cac = cac * (1 + cac_change_pct / 100)
        ltv = ltv * (1 + ltv_change_pct / 100)
        rev = rev * (1 + revenue_growth_pct / 100)

        ltv_cac = ltv / cac
        payback = (cac / (base_vals['aov_rs'] * (base_vals['gm_pct']/100) * base_vals['orders_yr'])) * 12
        # Payback: CAC inflation increases it linearly
        payback = payback * (1 + cac_change_pct / 100) ** (yr - 1) if yr > 0 else payback

        rows.append({
            'brand': brand_name,
            'fiscal_year': fy,
            'cac_rs': round(cac, 0),
            'ltv_rs': round(ltv, 0),
            'ltv_cac_ratio': round(ltv_cac, 3),
            'payback_period_months': round(payback, 1),
            'revenue_cr': round(rev, 1),
            'breach_flag': ltv_cac < 3.0,
        })
    return pd.DataFrame(rows)


# ── SCENARIO DEFINITIONS ──────────────────────────────────────────────────────
scenarios = {
    'Bear':  {'cac_change': +30, 'ltv_change': -5,  'rev_growth': +5},   # CAC +30% matches 2022-24 actual
    'Base':  {'cac_change': +15, 'ltv_change': +2,  'rev_growth': +12},  # Conservative moderation
    'Bull':  {'cac_change': -10, 'ltv_change': +8,  'rev_growth': +18},  # Retention + organic
}

all_projections = []

for brand in ['Mamaearth', 'boAt']:
    for scenario_name, params in scenarios.items():
        proj = project_scenario(
            brand_name=brand,
            base_vals=BASE[brand],
            cac_change_pct=params['cac_change'],
            ltv_change_pct=params['ltv_change'],
            revenue_growth_pct=params['rev_growth'],
            years=3
        )
        proj.insert(2, 'scenario', scenario_name)
        all_projections.append(proj)

projections = pd.concat(all_projections, ignore_index=True)
projections.to_csv("outputs/scenario_results.csv", index=False)
print("✅ Scenario projections complete")
print(f"   Shape: {projections.shape}")
print()

# Show summary
for brand in ['Mamaearth', 'boAt']:
    print(f"\n{'─'*55}")
    print(f"{brand} — LTV:CAC Ratio Projections")
    print(f"{'─'*55}")
    pivot = projections[projections['brand']==brand].pivot_table(
        index='scenario', columns='fiscal_year', values='ltv_cac_ratio'
    )
    pivot.columns.name = None
    pivot.index.name = None
    print(pivot.round(2).to_string())

    # Breach detection
    breaches = projections[(projections['brand']==brand) & (projections['breach_flag']==True)]
    if len(breaches) > 0:
        earliest = breaches.sort_values('fiscal_year').iloc[0]
        print(f"  🚨 BREACH: {brand} LTV:CAC < 3x in {earliest['fiscal_year']} under {earliest['scenario']} scenario")
    else:
        print(f"  ✅ No breach in any scenario within projection horizon (FY26-FY28)")
````

## Intervention ROI Model

**Three interventions modeled per brand:**
1. **CAC Reduction** — shift 15% of paid spend to organic/creator content
2. **AOV Increase** — bundling/upsell strategy, target +10-15% AOV lift
3. **Retention Improvement** — CRM/loyalty investment, reduce churn 10%, extend lifespan

````python
def model_intervention(brand_name, base_vals, intervention_type):
    """
    Model financial impact of a specific intervention.

    Returns: dict with cost_cr, ltv_cac_improvement, payback_improvement_months,
             net_revenue_impact_cr, roi_multiple
    """
    b = base_vals.copy()

    if intervention_type == 'cac_reduction':
        # Shift 15% of marketing spend from paid to organic/creator
        # Organic CAC is typically 30-40% of paid CAC
        shift_pct = 0.15
        spend_shifted_cr = b['mktg_cr'] * shift_pct
        organic_effectiveness_ratio = 0.40  # Organic acquires at 40% of paid cost

        # New customers from shifted spend (more efficient)
        new_cust_from_organic = (spend_shifted_cr * 1e7) / (b['cac_rs'] * organic_effectiveness_ratio)
        additional_cust_mn = new_cust_from_organic / 1e6

        # Net new CAC
        new_cac = b['cac_rs'] * (1 - shift_pct * (1 - organic_effectiveness_ratio))
        ltv_cac_new = b['ltv_rs'] / new_cac
        improvement = ltv_cac_new - (b['ltv_rs'] / b['cac_rs'])

        # Financial impact
        cust_increase_pct = additional_cust_mn / b['new_cust_mn']
        revenue_impact_cr = b['revenue_cr'] * cust_increase_pct * 0.5  # Conservative 50% conversion
        cost_cr = 0  # Reallocation, not incremental spend
        payback_improvement = (b['cac_rs'] - new_cac) / (b['aov_rs'] * (b['gm_pct']/100) * b['orders_yr']) * 12

        return {
            'intervention': 'CAC Reduction (Paid→Organic Shift)',
            'description': f'Shift {shift_pct*100:.0f}% of paid spend (₹{spend_shifted_cr:.0f}Cr) to creator/organic at 40% of paid CAC',
            'cost_cr': round(cost_cr, 1),
            'cac_before': round(b['cac_rs'], 0),
            'cac_after': round(new_cac, 0),
            'ltv_cac_before': round(b['ltv_rs'] / b['cac_rs'], 2),
            'ltv_cac_after': round(ltv_cac_new, 2),
            'ltv_cac_improvement': round(improvement, 2),
            'payback_improvement_months': round(payback_improvement, 1),
            'net_revenue_impact_cr': round(revenue_impact_cr, 1),
            'roi_multiple': round(revenue_impact_cr / max(cost_cr, 1), 1) if cost_cr > 0 else 999,
        }

    elif intervention_type == 'aov_increase':
        # Bundling / upsell — target +12% AOV
        aov_lift_pct = 0.12
        new_aov = b['aov_rs'] * (1 + aov_lift_pct)
        new_ltv = new_aov * b['orders_yr'] * (b['gm_pct']/100) * b['lifespan_yrs']
        ltv_cac_new = new_ltv / b['cac_rs']
        improvement = ltv_cac_new - (b['ltv_rs'] / b['cac_rs'])

        # Revenue impact: higher AOV × same customers
        revenue_impact_cr = b['revenue_cr'] * aov_lift_pct
        cost_cr = b['revenue_cr'] * 0.005  # ~0.5% investment in merchandising/tech

        payback_before = (b['cac_rs'] / (b['aov_rs'] * (b['gm_pct']/100) * b['orders_yr'])) * 12
        payback_after = (b['cac_rs'] / (new_aov * (b['gm_pct']/100) * b['orders_yr'])) * 12

        return {
            'intervention': 'AOV Increase (Bundling/Upsell)',
            'description': f'+{aov_lift_pct*100:.0f}% AOV via bundling; ₹{b["aov_rs"]:.0f} → ₹{new_aov:.0f} per order',
            'cost_cr': round(cost_cr, 1),
            'cac_before': round(b['cac_rs'], 0),
            'cac_after': round(b['cac_rs'], 0),
            'ltv_cac_before': round(b['ltv_rs'] / b['cac_rs'], 2),
            'ltv_cac_after': round(ltv_cac_new, 2),
            'ltv_cac_improvement': round(improvement, 2),
            'payback_improvement_months': round(payback_before - payback_after, 1),
            'net_revenue_impact_cr': round(revenue_impact_cr, 1),
            'roi_multiple': round(revenue_impact_cr / cost_cr, 1),
        }

    elif intervention_type == 'retention_improvement':
        # CRM/loyalty investment — reduce annual churn by 10% → extend lifespan by ~8 months
        churn_reduction = 0.10
        lifespan_extension_yrs = 0.65  # ~8 months additional lifespan
        new_lifespan = b['lifespan_yrs'] + lifespan_extension_yrs
        new_ltv = b['aov_rs'] * b['orders_yr'] * (b['gm_pct']/100) * new_lifespan
        ltv_cac_new = new_ltv / b['cac_rs']
        improvement = ltv_cac_new - (b['ltv_rs'] / b['cac_rs'])

        # Investment: CRM platform + 1-2 retention campaigns per quarter
        cost_cr = b['revenue_cr'] * 0.015  # 1.5% of revenue = standard CRM/loyalty budget

        # Revenue impact: retained customers × AOV × orders
        retained_cust_mn = b['new_cust_mn'] * churn_reduction
        revenue_impact_cr = retained_cust_mn * 1e6 * b['aov_rs'] * b['orders_yr'] / 1e7

        return {
            'intervention': 'Retention Improvement (CRM/Loyalty)',
            'description': f'10% churn reduction; lifespan {b["lifespan_yrs"]:.1f}yrs → {new_lifespan:.1f}yrs; invest ₹{cost_cr:.0f}Cr in CRM',
            'cost_cr': round(cost_cr, 1),
            'cac_before': round(b['cac_rs'], 0),
            'cac_after': round(b['cac_rs'], 0),
            'ltv_cac_before': round(b['ltv_rs'] / b['cac_rs'], 2),
            'ltv_cac_after': round(ltv_cac_new, 2),
            'ltv_cac_improvement': round(improvement, 2),
            'payback_improvement_months': 0,  # Payback unaffected
            'net_revenue_impact_cr': round(revenue_impact_cr, 1),
            'roi_multiple': round(revenue_impact_cr / cost_cr, 1),
        }


# ── RUN ALL INTERVENTIONS ──────────────────────────────────────────────────────
intervention_types = ['cac_reduction', 'aov_increase', 'retention_improvement']
intervention_results = []

for brand in ['Mamaearth', 'boAt']:
    for itype in intervention_types:
        result = model_intervention(brand, BASE[brand], itype)
        result['brand'] = brand
        intervention_results.append(result)

df_interventions = pd.DataFrame(intervention_results)
df_interventions.to_csv("outputs/intervention_results.csv", index=False)

print("✅ Intervention ROI model complete")
print()
for brand in ['Mamaearth', 'boAt']:
    print(f"\n{'─'*65}")
    print(f"{brand} — Intervention ROI Summary")
    print(f"{'─'*65}")
    df_ = df_interventions[df_interventions['brand']==brand][
        ['intervention','cost_cr','ltv_cac_before','ltv_cac_after',
         'ltv_cac_improvement','net_revenue_impact_cr','roi_multiple']
    ].reset_index(drop=True)
    print(df_.to_string(index=False))
````

````python
methodology = """
# METHODOLOGY — D2C Unit Economics Stress-Test
## Mamaearth vs boAt | FY22–FY25 Analysis + FY26–FY28 Projections

---

## 1. DATA SOURCES

### Mamaearth (Honasa Consumer Ltd — BSE: 543832)
- **Primary:** BSE Annual Reports FY22–FY25 (Audited P&L, Balance Sheet, Cash Flow)
- **Secondary:** Investor Presentations (BSE Corporate Announcements tab)
- **Cross-check:** ICICI Securities Initiation Coverage Report (available via BSE analyst reports)
- **Earnings commentary:** Q4FY25 Earnings Call Transcript — operating profit margin disclosure

### boAt (Imagine Marketing Ltd)
- **Primary:** SEBI DRHP (October 2025) — Restated 3-year financials (FY23-FY25)
- **FY25 standalone:** Company press release — ₹3,070 Crore revenue, 4.64% EBITDA margin
- **FY22:** Estimated from public filings and press releases prior to DRHP

### Industry Benchmarks
- CAC inflation ~30% YoY: Unicommerce D2C Report 2024; Redseer Strategy Consultants
- LTV:CAC 3x threshold: Standard referenced across Indian D2C analyst coverage
- Market size: IMARC Group India D2C Forecast 2025

---

## 2. METRIC DERIVATIONS

### CAC (Customer Acquisition Cost)
**Formula:** `CAC (₹) = Marketing Spend (₹Cr × 10M) / New Customers (Mn × 1M)`

**Confidence: MEDIUM**

**New Customers Estimation:**
Neither brand directly discloses "new customers acquired." We derive it as:
`New Customers (t) = Active Consumers (t) - Active Consumers (t-1) + Estimated Churn (t-1)`

Churn estimates:
- Mamaearth: 18–20% annual churn (based on 44% repeat rate; multi-brand offsets pure single-brand churn)
- boAt: 25–28% annual churn (hardware replacement cycle; 28% repeat rate in FY25)

**Assumption documented:** All marketing spend (A&P line in P&L) is attributed to new customer acquisition.
In reality, ~15-25% of marketing spend serves retention and brand purposes. This makes our CAC estimates
conservative (slightly higher than actual). This is the correct direction for stress-testing — we are
testing against a conservative CAC, not an optimistic one.

**Cross-check:** Derived Mamaearth FY24 CAC of approximately ₹1,900-2,100 is consistent with
analyst commentary in ICICI Securities coverage report range.

---

### LTV (Customer Lifetime Value)
**Formula:** `LTV (₹) = AOV × Orders/Year × Gross Margin % × Customer Lifespan (years)`

**Confidence: MEDIUM**

**AOV Derivation:** Revenue / Estimated Total Orders
- Total Orders = Active Consumers × Avg Orders/Year

**Orders/Year:** Estimated from disclosed repeat purchase rate (investor presentations).
A 44% annual repeat rate ≈ 2.7-2.8 orders/year for a brand with multi-SKU purchases.

**Gross Margin %:** Directly calculated from reported P&L: (Revenue - COGS) / Revenue.
Gross profit is disclosed in annual reports. **Confidence: HIGH**

**Customer Lifespan:**
- Mamaearth: 2.4–2.5 years. Based on: India beauty D2C benchmark 2-3 years;
  Mamaearth's multi-brand strategy (The Derma Co., Aqualogica) extends lifespan above single-brand.
  **Confidence: LOW-MEDIUM** — most sensitive assumption in LTV model.
- boAt: 3.0–3.5 years. Hardware ecosystem creates longer cycles; replacement + accessories.
  **Confidence: LOW-MEDIUM**

**Sensitivity note:** A 6-month change in customer lifespan assumption changes LTV:CAC ratio by ~0.3-0.5x.
This is why we present scenario ranges rather than point estimates.

---

### LTV:CAC Ratio
**Formula:** `LTV:CAC = LTV / CAC`
**Threshold:** 3.0x = industry minimum for sustainable D2C operations
**Confidence:** Derived — inherits confidence levels of both LTV and CAC

---

### Payback Period
**Formula:** `Payback (months) = [CAC / (AOV × Gross Margin % × Orders/Year)] × 12`
This equals: months to recover CAC from gross profit contribution per customer per year.
**Confidence: MEDIUM** (same driver uncertainty as CAC)

---

### Marketing Efficiency Ratio
**Formula:** `Revenue / Marketing Spend`
This is a revenue-ROAS proxy. Higher = more revenue generated per rupee of ad spend.
**Confidence: HIGH** — both inputs are directly reported.

---

## 3. SCENARIO MODEL ASSUMPTIONS

| Parameter | Bear | Base | Bull | Justification |
|-----------|------|------|------|---------------|
| CAC YoY Change | +30% | +15% | -10% | Bear = documented 2022-24 actual; Base = moderated; Bull = organic channel push |
| LTV YoY Change | -5% | +2% | +8% | Bear = AOV pressure + shorter lifespan; Bull = retention investment |
| Revenue Growth | +5% | +12% | +18% | Correlated with customer efficiency scenario |
| Projection Horizon | 3 years (FY26-FY28) | Same | Same | Sufficient to show breach timeline |

---

## 4. INTERVENTION MODEL ASSUMPTIONS

### CAC Reduction (Paid→Organic Shift)
- 15% of marketing spend shifted to organic/creator content
- Organic CAC = 40% of paid CAC (conservative; industry range 30-50%)
- No incremental cost (reallocation)

### AOV Increase (Bundling/Upsell)
- 12% AOV lift target (industry achievable through bundling)
- Investment = 0.5% of revenue in merchandising/tech development
- Conservative: not all customers adopt bundles

### Retention Improvement (CRM/Loyalty)
- 10% annual churn reduction
- Lifespan extension: ~8 months (from 10% × ~6.5yr = 0.65yr added)
- Investment = 1.5% of revenue (standard CRM + loyalty program budget)

---

## 5. LIMITATIONS AND KNOWN UNCERTAINTIES

1. **CAC is derived, not disclosed** — primary source of model uncertainty
2. **LTV lifespan assumption** — sensitive; verified against industry benchmarks but not brand-specific data
3. **boAt FY22 estimates** — derived from press releases, not DRHP (covers only FY23-FY25)
4. **Attribution of marketing spend** — assumes 100% acquisition-directed; likely 10-20% retention
5. **Scenario projections** — linear extrapolation; real outcomes are nonlinear

All limitations are conservative in direction: they make our stress-test more demanding, not less.
If the brand can maintain LTV:CAC > 3x even under these conservative derivations, the underlying
economics are likely stronger than our model suggests.

---

## 6. DATA INTEGRITY VERIFICATION

- Mamaearth revenue FY25 (₹2,066Cr): Consistent with announced results
- Mamaearth operating profit margin FY25 (3.3%): Directly from Q4FY25 earnings call
- boAt EBITDA margin FY25 (4.64%): Directly from company press release
- boAt channel split (70.55% online / 29.45% offline): Directly from DRHP disclosure
- Gross margins: Calculated directly from reported P&L; no derivation required

"""

with open("docs/methodology.md", "w") as f:
    f.write(methodology)
print("✅ docs/methodology.md written")
print(f"   Length: {len(methodology)} characters")
````
